```python
import pyrealsense2 as rs
import numpy as np
import csv

# === File settings ===
bag_file = "20250611_115438.bag"
output_csv = "normal2.csv"

# === Set up pipeline ===
pipeline = rs.pipeline()
config = rs.config()
config.enable_device_from_file(bag_file, repeat_playback=False)
config.enable_stream(rs.stream.depth, 848, 480, rs.format.z16, 10)

pipeline.start(config)

# Get playback object to check total frames (optional)
playback = pipeline.get_active_profile().get_device().as_playback()
playback.set_real_time(False)  # Speeds up playback

try:
    frame_count = 0
    with open(output_csv, 'w', newline='') as file:
        writer = csv.writer(file)

        while True:
            frames = pipeline.wait_for_frames()
            depth_frame = frames.get_depth_frame()
            if not depth_frame:
                continue

            # Convert depth to 2D numpy array
            depth_image = np.asanyarray(depth_frame.get_data())
            height, width = depth_image.shape

            # Write one frame to CSV as flat row (optional reshape)
            writer.writerow(depth_image.flatten())
            frame_count += 1
            print(f"Saved frame {frame_count}")

except Exception as e:
    print("Error:", e)

finally:
    pipeline.stop()
    print("Done. CSV saved:", output_csv)
```
